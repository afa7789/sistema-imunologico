```python
import numpy as np
import random
import math
```

```python
def criarPopulacao(tamanho):
    listaRandomX=[]
    listaRandomY=[]
    for i in range(0,tamanho):
        listaRandomX.append(random.uniform(-10,10));
        listaRandomY.append(random.uniform(-10,10));
    arrayX = np.array(listaRandomX)
    arrayY = np.array(listaRandomY)
    return arrayX,arrayY
```
